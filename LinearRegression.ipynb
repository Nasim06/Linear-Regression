```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

Task 3.1

```python
y = np.load('y_values.npy')
x = np.load('x_values.npy')
testX = np.load('test_x_values.npy')

reg = LinearRegression().fit(x, y)
prediction = reg.predict(testX)

plt.scatter(x, y)
plt.plot(testX, prediction)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.show()

m = reg.coef_
c = reg.intercept_
y = c + (m * 0.48)
print(y)


```

Task 3.2

```python
WineX = np.load('WineData.npy')
WineY = np.load('WineLabels.npy')

n = WineX.shape[0]
split_point = int(n * 0.8)

train_data = WineX[:split_point, :] 
train_labels = WineY[:split_point] 
test_data = WineX[split_point:, :] 
test_labels = WineY[split_point:] 

plt.scatter(train_data[:, 0], train_data[:, 1], marker='o', c=train_labels)
plt.scatter(test_data[:, 0], test_data[:, 1], marker='^', c=test_labels)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()

```

```python
pca = PCA(n_components=2)
tTrain_data = pca.fit_transform(train_data)
tTest_data = pca.transform(test_data)

plt.scatter(tTrain_data[:, 0], tTrain_data[:, 1], marker='o', c=train_labels)
plt.scatter(tTest_data[:, 0], tTest_data[:, 1], marker='x', c=test_labels)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()

```

Task 3.3

```python
lda = LDA(n_components=2)
lda.fit(train_data, train_labels)
LTrain_data = lda.transform(train_data)
LTest_data = lda.transform(test_data)

plt.scatter(LTrain_data[:, 0], LTrain_data[:, 1], marker='o', c=train_labels)
plt.scatter(LTest_data[:, 0], LTest_data[:, 1], marker='x', c=test_labels)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()
```
